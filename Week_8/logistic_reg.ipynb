```python
import pandas as pd
df = pd.read_csv('D:\Python\TekWorks\Week_8\customer_churn.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['income'] = df['income'].fillna(df['income'].mode()[0])
```

```python
df['credit_score'] = df['credit_score'].fillna(df['credit_score'].mean())
```

```python
df['avg_purchase_value'] = df['avg_purchase_value'].fillna(df['avg_purchase_value'].mean())
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
q1 = df['income'].quantile(0.80)
q2 = df['income'].quantile(0.20)
df['income'] = df['income'].clip(lower=q2,upper=q1)
```

```python
q1 = df['avg_purchase_value'].quantile(0.80)
q2 = df['avg_purchase_value'].quantile(0.20)
df['avg_purchase_value'] = df['avg_purchase_value'].clip(lower=q2,upper=q1)
```

```python
q1 = df['days_since_last_login'].quantile(0.80)
q2 = df['days_since_last_login'].quantile(0.20)
df['days_since_last_login'] = df['days_since_last_login'].clip(lower=q2,upper=q1)
```

```python
df.describe()
```

```python
df.head()
```

```python
df['gender'] = df['gender'].replace({'Male':0,'Female':1})
```

```python
df.region.unique()
```

```python
df['region'] = df['region'].replace({'North':0,'South':1,'East':2,'West':3})
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
high_scale = ['income', 'credit_score', 'avg_purchase_value']
df[high_scale] = pd.DataFrame(scaler.fit_transform(df[high_scale]), columns=high_scale, index=df.index)

```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , precision_score , confusion_matrix
accuracy_score(y_pred,y_test)*100
```

```python
confusion_matrix(y_pred,y_test)
```
