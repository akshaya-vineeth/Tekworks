```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.amazon.in/s?k=laptops&crid=831M2GKDLEMO&sprefix=lapto%2Caps%2C346&ref=nb_sb_noss_2"
```

```python
res = requests.get(url)
res.text
```

```python
soup = BeautifulSoup(res.content,"html.parser")
# Check if h1 exists before accessing text
h1_tag = soup.find('h1')
if h1_tag:
    print(h1_tag.text.strip())
else:
    print("No h1 tag found")
```
