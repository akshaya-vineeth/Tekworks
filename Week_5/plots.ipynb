```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
a = pd.read_csv('D:\Python\TekWorks\Week_5\dataset.csv')
```

```python
df = pd.DataFrame(a)
df.head()
```

# 1. HISTOGRAM

```python
plt.figure(figsize=(10,6))
sns.histplot(df['age'],kde=True,color='pink',bins=5,alpha=0.7)
plt.xlabel("AGE")
plt.ylabel("COUNT")
plt.show()
```

# 2. SCATTER PLOT


```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=df['trestbps'],y=df['chol'],data=df,color='black',edgecolor='black')
plt.xlabel("TrestBPS")
plt.ylabel("CHOLESTROL")
```

# 3. LINE PLOT

```python
sns.lineplot(x='age',y='chol',data=df,color='green',markers='o')
plt.grid(True)
```

# 4. BOX PLOT

```python
sns.boxplot(x='age',data=df,color='orange')
```

```python
# 5. PIE CHART
target_counts = df['target'].value_counts()
colors = ['#FF6B6B', '#4ECDC4']
explode = (0.05, 0.05)

plt.figure(figsize=(10, 8), facecolor="#17292F")
ax = plt.gca()
ax.set_facecolor("#0E232A")
plt.pie(target_counts, 
        labels=['Heart Disease Present', 'No Heart Disease'],
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        explode=explode,
        shadow=True,
        textprops={'fontsize': 12, 'weight': 'bold'})

plt.title('Distribution of Heart Disease Cases', fontsize=14, weight='bold', pad=20)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
