```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC , SVR
from sklearn.metrics import classification_report , accuracy_score , confusion_matrix ,r2_score
from sklearn.preprocessing import StandardScaler

```

```python

df = pd.read_csv("C:\\Users\\vinee\Downloads\\heart-2.csv")
df.head()
```

```python
df.isna().sum()
```

```python
def model(df):
    X = df.iloc[: , :-1]
    y = df.iloc[: , -1]
    X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2)
    sc = StandardScaler()
    X_train_scaled = sc.fit_transform(X_train)
    X_test_scaled = sc.transform (X_test)
    #lg = ()
    #lg.fit(X_train , y_train)
    svm = SVC(kernel='linear',probability=True,C=1,max_iter=100,break_ties=True,random_state=42)
    svm.fit(X_train_scaled,y_train)
    lg_preds = svm.predict(X_test)
    print(classification_report(y_test , lg_preds))
    cm = confusion_matrix(y_test , lg_preds)
    print(cm)
```

```python
model(df)
```
