1. Why and What is random forest(base learners)?
    --> To overcome the overfitting problem from decision tree.
    --> It is a combination of multiple decision trees and it is a bagging technique.
    --> In decision tree there will be overfitting problem with low bias and high variance, but in random forest it is 
    reduced because the decision or the prediction is taken combining no of decision trees so the variance is reduced.
    --> The decision trees are connected parallely.

2. Intro to ensemble learning?
    --> In ensemble technique we have bagging and boosting.
    --> It is a combination of two or more decision trees to make a prediction.

3. Boosting?(weak learners-nothing much is learnt from dataset)
    1. adaboost?
        --> in this the decision trees are connected sequentially and the weights or the losses are sent to next decision tree.
        --> as all the weak learners are combined to get a strong learner or strong decision.
        --> formula : a1(b1) + a2(b2) + a3(b3) + ...... + an(bn)
            b1....n ---> decision tree steps or no of decision trees
            a1....n ---> weights



4. Stacking or Voting?
    --> It is the process of applying all classification or regression algo's and see the best algo and select it accuracy.
