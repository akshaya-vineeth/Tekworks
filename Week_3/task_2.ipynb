```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("D:\Python\TekWorks\Week_3\Dataset_1.csv")
df = pd.DataFrame(a)
```

```python
df.head()
```

```python
#1
df['order_amount'] = df['order_amount'].replace('₹999',999,inplace=True)
df['order_amount'] = df['order_amount'].astype('float')
df['order_amount'].unique()
```

```python

df['is_delivered'] = df['is_delivered'].replace('yes','True',inplace=True)
df['is_delivered'] = df['is_delivered'].replace('no','False',inplace=True)
df['is_delivered'].unique()


```

```python
df.query("order_amount.isna() and is_delivered == 'True'").head()
```

```python
#2
df['order_date'] = df['order_date'].replace('invalid',np.nan)
df['order_date'].unique()
```

```python

df['order_date'] = pd.to_datetime(df['order_date'],format='mixed',errors='coerce',dayfirst=True)
```

```python
df['delivery_days'] =df['delivery_days'].replace('three','3',inplace=True)
df['delivery_days'] =df['delivery_days'].astype('float64')
```
